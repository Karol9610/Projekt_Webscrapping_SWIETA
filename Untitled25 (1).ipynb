```python
import pandas as pd
import requests  
from bs4 import BeautifulSoup # PARSE HTML
```

```python
value = '1999'  #Rok w którym występują święta --> tutaj powinna być funlcja z modelera to_year(Datetime2)
```

```python
wikiurl=f"https://www.kalendarz-365.pl/swieta/{value}.html" #ADRES URL STRONY
table_class="full-view-table table table-striped table-hover tablesorter table-small-font tablesorter-default sort-v1" #ZBADAJ STRONĘ, POSZUKAJ TABELI(KLASY)
response=requests.get(wikiurl) #ad1
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

tabela_swieta=soup.find('table',{'class':"full-view-table"})

df=pd.read_html(str(tabela_swieta)) #LISTA
df=pd.DataFrame(df[0]) #DATAFRAME
```

```python
df
```

```python
data = df.drop([ "Dni jeszcze"], axis=1) # usunięcie kolumny
data = data.rename(columns={"Święto": "Swieto","Dzień": "Dzien"}) #zmiana nazwy kolumn 
```

```python
data["Data"]= data["Data"].replace({     # zmiana nazw_miesięcy (słownie) na misiące (numerycznie)
    ' stycznia ': '.01.', 
    ' lutego ': '.02.',
    ' marca ': '.03.', 
    ' kwietnia ': '.04.', 
    ' maja ': '.05.', 
    ' czerwca ': '.06.', 
    ' lipca ': '.07.', 
    ' sierpnia ': '.08.', 
    ' września ': '.09.', 
    ' października ': '.10.', 
    ' listopada ': '.11.', 
    ' grudnia ': '.12.'},
     regex=True)
```

```python
data
```

```python
filename = f"{value}_swieta.csv"
data.to_csv(filename, sep='\t', encoding='utf-8') #Plik zapisuje się tam gdzie inne pliki Jupyter notebook
```
